```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\my pc\Desktop\case analytics 2\Enigma Case Study_Updated\Enigma_Updated_Data Set.xlsx')
df.head()
```

```python
cols_at_end = ['Order total']
df = df[[c for c in df if c not in cols_at_end] 
        + [c for c in cols_at_end if c in df]]
```

```python
df = df[['Customer placed order datetime','Restaurant ID','SuperOrder','Amount of tip','Delivery Region','Order total']]
```

```python
df.tail()
```

```python
df.dtypes
```

```python

df['Date'] =df['Customer placed order datetime'].str[:2]
```

```python
df['Customer placed order datetime'] =df['Customer placed order datetime'].str[3:]
```

```python
df.head()
```

```python
df.loc[(df['Customer placed order datetime'] >"06:00:00") & (df['Customer placed order datetime'] <="12:00:00"), 'Time'] = 'MORNING' 
df.loc[(df['Customer placed order datetime'] >"12:00:00") & (df['Customer placed order datetime'] <="18:00:00"), 'Time'] = 'AFTERNOON' 
df.loc[(df['Customer placed order datetime'] >"18:00:00") & (df['Customer placed order datetime'] <="24:00:00"), 'Time'] = 'LATE NIGHT' 
df.loc[(df['Customer placed order datetime'] >"00:00:00") & (df['Customer placed order datetime'] <="06:00:00"), 'Time'] = 'EARLY MORNING' 
 
```

```python
df
```

```python
df1 = df.drop(['Customer placed order datetime'],axis=1)
```

```python
dummiesregion = pd.get_dummies(df1['Delivery Region'])
dummiestime = pd.get_dummies(df1['Time'])
dummiessuperorder = pd.get_dummies(df1['SuperOrder'])
```

```python
df2 = pd.concat([df1,dummiesregion,dummiestime,dummiessuperorder],axis=1)
```

```python
df2.head()
```

```python
X=df2.drop(['Order total','Delivery Region','Time','SuperOrder'], axis=1)
Y=df2['Order total']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
##create a pickle file
import pickle
pickle_out =open('regressor.pkl','wb')
pickle.dump(regressor,pickle_out)
pickle_out.close()
```
